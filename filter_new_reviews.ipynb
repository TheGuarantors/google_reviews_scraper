```python
import pandas as pd
import ast
```

```python
t1 = pd.read_csv('1.csv')
t2 = pd.read_csv('2.csv')
t3 = pd.read_csv('3.csv')
t4 = pd.read_csv('4.csv')
t5 = pd.read_csv('5.csv')
t6 = pd.read_csv('6.csv')
t7 = pd.read_csv('7.csv')
t8 = pd.read_csv('8.csv')
```

```python
t8
```

```python
data = t8.loc[t8['reviews'].notnull()].reset_index()
```

```python
type(data)
```

```python
data
```

```python
final_df = pd.DataFrame(columns = ["address", "star1_cnt", "star2_cnt", "star3_ct", "star4_cnt", "star5_cnt", "with_text_cnt", "total_cnt"])

for row in range(len(data)):
    str_json = data['reviews'].tolist()[row]
    building_dict = ast.literal_eval(str_json)
    #iterating over dictionary
    star1 = 0
    star2 = 0
    star3 = 0
    star4 = 0
    star5 = 0
    with_text = 0
    total = 0
    for key, value in building_dict.items():
        #will be extracted only if review is sufficietly old:
        if 'year' in value['Date']:
            if value['Rating'] == '1':
                star1 += 1
            elif value['Rating'] == '2':
                star2 += 1
            elif value['Rating'] == '3':
                star3 += 1  
            elif value['Rating'] == '4':
                star4 += 1
            elif value['Rating'] == '5':
                star5 += 1
            if value['Text'] != '':
                with_text += 1
            total += 1
    new_row = pd.DataFrame([[data['Unnamed: 0'][row], star1, star2, star3, star4, star5, with_text, total]],
                      columns = ["address", "star1_cnt", "star2_cnt", "star3_ct", "star4_cnt", "star5_cnt", "with_text_cnt", "total_cnt"])
    final_df = final_df.append(new_row)
```

```python
final_df.columns = map(str.upper, final_df.columns)
final_df
```

```python
#final step
final_df.to_csv('GOOGLE_REVIEWS_1_YEAR_OLD.csv', index = False)
```
